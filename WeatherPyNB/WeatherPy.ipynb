```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import requests
import time
import json
from scipy.stats import linregress
import openweathermapy as owm
from pprint import pprint

# Import API key
from weather_key import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)


```

#### Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
query_url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key
```

```python
city_name = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []
 
 
for city in cities:  
    
    
    try: 
        response = requests.get(f"{query_url}&q={city}").json() 
        city_name.append(response["name"])
        cloudiness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        max_temp.append(response["main"]["temp_max"])
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        wind_speed.append(response["wind"]["speed"])
        city_record = response["name"]
        print(f"{city_record}")
        print(f"{query_url}&q={city}")
        
    except:
        print("City not found")
    continue
```

###### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_dic = {"City": city_name,
                          "Cloudiness": cloudiness,
                          "Country": country,
                          "Date": date,
                          "Humidity": humidity,
                          "Max Temperature" : max_temp,
                          "Latitude": lat,
                          "Longitude": lng,
                          "Wind Speed": wind_speed
                          }
weather_df = pd.DataFrame(weather_dic, columns = ['City','Cloudiness', 'Country', 'Date', 'Humidity',
                                                'Max Temperature', 'Latitude', 'Longitude', 'Wind Speed'])
weather_df["Date"] = pd.to_datetime(weather_df["Date"], infer_datetime_format=True, format="%Y/%m/%d")
'{:.0%}'



weather_df.to_csv('weatherdata.csv', index=False, header=True)
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x = northern_df['Max Temperature']
y = northern_df['Latitude']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x,y)
regress = x * slope + intercept
line_eq = f"y = {round(slope, 2)} x + {round(intercept, 2)}"
plt.scatter(x, y)
plt.plot(x,regress,color = 'black', linewidth= 3)
plt.annotate(line_eq,(100,75), color='red', fontsize=18, ha= 'center')
plt.xlabel("Max Temperature")
plt.ylabel("Latitude")
plt.title("Northern Hemisphere - Max Temp vs Latitude")
plt.style.use(['seaborn-pastel'])
plt.savefig("NH_Temp_vs_Lat.png")
plt.show()
```
